```python
person={"name":"adnan","age":19}
print(person["age"])
```

```python
# 3. Add a new key-value pair to a dictionary.
# Question: Add a key "city" with value "New York" to the dictionary person.
```

```python
person={"name":"adnan","age":19}
person["city"]="delhi"
print(person)
```

```python
# 4. Update an existing value in a dictionary.
# Question: Change the age of "Alice" to 28 in the given dictionary.
```

```python
person={"name":"adnan","age":19}
person["age"]=18
print(person)
```

```python
# 5. Delete a key from a dictionary.
# Question: Remove the "age" key from the dictionary.
```

```python
person={"name":"adnan","age":19}
del person["age"]
print(person)
```

```python
# 6. Get the length of a dictionary.
# Question: Find how many items are in the dictionary.
```

```python
person={"name":"adnan","age":19}
print(len(person))
```

```python
# 7. Copy a dictionary.
# Question: Create a copy of the dictionary person.
```

```python
person={"name":"adnan","age":19}
copy_person=person.copy()
print(copy_person)
```

```python
# 8. Use get() method to access value safely.
# Question: Access the "city" key using get(), provide default "Unknown" if not present.
```

```python
person={"name":"adnan","age":19}
city=person.get("city","unknown")
print(person)
```
